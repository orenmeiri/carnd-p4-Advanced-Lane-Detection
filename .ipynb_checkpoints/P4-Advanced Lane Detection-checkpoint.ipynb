# Camera Calibration

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pickle
from moviepy.editor import VideoFileClip
import scipy
```

```python
objpoints = []
imgpoints = []

objp = np.zeros((9*6,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

for i in range(1, 21):
    imgname = 'camera_cal/calibration{}.jpg'.format(i)
    img = mpimg.imread(imgname)
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    ret, corners = cv2.findChessboardCorners(gray, (9,6),None)
    if ret:
#         img = cv2.drawChessboardCorners(img, (9,6), corners, ret)
        imgpoints.append(corners)
        objpoints.append(objp)
    else:
        print("failed to find corners in {}".format(imgname))
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img.shape[0:2],None,None)

# Save the camera calibration result for later use (we won't worry about rvecs / tvecs)
dist_pickle = {}
dist_pickle["mtx"] = mtx
dist_pickle["dist"] = dist
pickle.dump( dist_pickle, open( "dist_pickle.p", "wb" ) )
```

# Thresholding

```python
def abs_sobel_thresh(gray, orient='x', sobel_kernel=3, thresh=(0, 255)):
    if orient == 'x':
        sobel = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
    else:
        sobel = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
    abs_sobel = np.absolute(sobel)
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    sbinary = np.zeros_like(scaled_sobel)
    sbinary[(scaled_sobel >= thresh[0]) & (scaled_sobel <= thresh[1])] = 1
    return sbinary

def mag_thresh(gray, sobel_kernel=3, thresh=(0, 255)):
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, sobel_kernel)
    sobel_magnitude = np.sqrt(sobelx**2 + sobely ** 2)
    scaled_sobel = np.uint8(255*sobel_magnitude/np.max(sobel_magnitude))
    sbinary = np.zeros_like(scaled_sobel)
    sbinary[(scaled_sobel >= thresh[0]) & (scaled_sobel <= thresh[1])] = 1
    # binary_output = np.copy(img) # Remove this line
    return sbinary

def dir_threshold(gray, sobel_kernel=3, thresh=(0, np.pi/2)):
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    abs_sobelx = np.absolute(sobelx)
    abs_sobely = np.absolute(sobely)
    direction = np.arctan2(abs_sobely, abs_sobelx)
    sbinary = np.zeros_like(gray)
    sbinary[(direction >= thresh[0]) & (direction <= thresh[1])] = 1
    return sbinary
```

```python
def analyze(name, image):
    undistorted = cv2.undistort(image, mtx, dist, None, mtx)
    hls = cv2.cvtColor(undistorted, cv2.COLOR_RGB2HLS)
    H = hls[:,:,0]
    L = hls[:,:,1]
    S = hls[:,:,2]
    clean_S = np.zeros_like(S)
    clean_S[(S >= 90) & (S <= 255)] = 1
    clean_L = np.zeros_like(L)
    clean_L[(L >= 120) & (L <= 255)] = 1
    combined = np.minimum(L, S)
    mean_S = np.mean(S)
    mean_L = np.mean(L)
    gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
    mag = mag_thresh(gray, sobel_kernel=15, thresh=(30, 100))
    direction = dir_threshold(gray, sobel_kernel=15, thresh=(0.7, 1.3))
    mag_direction = mag & direction
    if mean_S > 70:
        output = clean_L | mag_direction
    elif mean_L > 115:
        output = clean_S | mag_direction
    else:
        output = clean_S | clean_L | mag_direction
    print("{} mean S {} L {}".format(name, np.mean(S), np.mean(L)))
    return undistorted, S, clean_S, L, clean_L, mag_direction, output

image_names = [
    "left",
    "right",
    "straight",
    "white",
    "white1",
    "white2",
    "challenge1",
    "bad_image1",
    "bad_image2"
              ]

f, axarr = plt.subplots(len(image_names), 7, figsize=(24, len(image_names) * 4))
axarr[0,0].set_title('undistorted')
axarr[0,1].set_title('S')
axarr[0,2].set_title('clean_S')
axarr[0,3].set_title('L')
axarr[0,4].set_title('clean_L')
axarr[0,5].set_title('mag_direction')
axarr[0,6].set_title('output')
for idx, name in enumerate(image_names):
    axarr[idx, 0].set_title(name)
    image_name = "{}.jpg".format(name)
    image = mpimg.imread(image_name)
    img1, img2, img3, img4, img5, img6, img7 = analyze(name, image)
    axarr[idx, 0].imshow(img1)
    axarr[idx, 1].imshow(img2, cmap='gray')
    axarr[idx, 2].imshow(img3, cmap='gray')
    axarr[idx, 3].imshow(img4, cmap='gray')
    axarr[idx, 4].imshow(img5, cmap='gray')
    axarr[idx, 5].imshow(img6, cmap='gray')
    axarr[idx, 6].imshow(img7, cmap='gray')
plt.show()
# plt.savefig('output_images/binary_analysis.jpg')
```

```python
def preprocess_img(undistorted):
    hls = cv2.cvtColor(undistorted, cv2.COLOR_RGB2HLS)
    H = hls[:,:,0]
    L = hls[:,:,1]
    S = hls[:,:,2]
    clean_S = np.zeros_like(S)
    clean_S[(S >= 90) & (S <= 255)] = 1
    clean_L = np.zeros_like(L)
    clean_L[(L >= 120) & (L <= 255)] = 1
    combined = np.minimum(L, S)
    mean_S = np.mean(S)
    mean_L = np.mean(L)
    gray = cv2.cvtColor(undistorted,cv2.COLOR_RGB2GRAY)
    mag = mag_thresh(gray, sobel_kernel=15, thresh=(30, 100))
    direction = dir_threshold(gray, sobel_kernel=15, thresh=(0.7, 1.3))
    mag_direction = mag & direction
    if mean_S > 70:
        output = clean_L | mag_direction
    elif mean_L > 115:
        output = clean_S | mag_direction
    else:
        output = clean_S | clean_L | mag_direction
#     print("{} mean S {} L {}".format(name, np.mean(S), np.mean(L)))
    return output
```

# Perspective Transform

```python
def warp(img, M):
    img_size = (img.shape[1], img.shape[0])
    warped = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
    return warped

def unwarp(img, Minv):
    img_size = (img.shape[1], img.shape[0])
    unwarped = cv2.warpPerspective(img, Minv, img_size) 
    return unwarped
```

```python
def draw_lines(img, points, color=[255, 0, 0], thickness=5):
    output = img.copy()
    for i, point in enumerate(points):
        point_next = points[(i+1)%(len(points))]
        cv2.line(output, tuple(point), tuple(point_next), color, thickness)
    return output
        
```

```python
pnt1 = [30, 720]
pnt2 = [550, 465]
pnt3 = [730, 465]
pnt4 = [1250, 720]

points = [pnt1, pnt2, pnt3, pnt4]
src = np.array(points, np.float32)
dest = np.array([[0, 720], [0, 0], [1280, 0], [1280, 720]], np.float32)
M = cv2.getPerspectiveTransform(src, dest)

image = mpimg.imread("straight.jpg")
undistorted = cv2.undistort(image, mtx, dist, None, mtx)
lines = draw_lines(undistorted, points = points)
plt.imshow(lines)
plt.show()
# plt.savefig('output_images/straight_transform_src.jpg')

warped = warp(lines, M)
plt.imshow(warped)
plt.show()
# plt.savefig('output_images/straight_transform_dest.jpg')
```

# Finding the lane marks

```python
# Assuming you have created a warped binary image called "binary_warped"

class Line():
    def __init__(self):
        # was the line detected in the last iteration?
        self.detected = False  
        # x values of the last n fits of the line
        self.recent_xfitted = [] 
        #average x values of the fitted line over the last n iterations
        self.bestx = None     
        #polynomial coefficients averaged over the last n iterations
        self.best_fit = None  
        #polynomial coefficients for the most recent fit
        self.current_fit = [np.array([False])]  
        #radius of curvature of the line in some units
        self.radius_of_curvature = None 
        #distance in meters of vehicle center from the line
        self.line_base_pos = None 
        #difference in fit coefficients between last and new fits
        self.diffs = np.array([0,0,0], dtype='float') 
        #x values for detected line pixels
        self.allx = None  
        #y values for detected line pixels
        self.ally = None
        
def detect_lanes(binary_image):
    global left_line
    global right_line
    global ym_per_pix
    global xm_per_pix
    # Take a histogram of the bottom half of the image
    ploty = np.linspace(0, binary_image.shape[0]-1, binary_image.shape[0] )
    histogram = np.sum(binary_image[int(binary_image.shape[0]/2):,:], axis=0)
    # Create an output image to draw on and  visualize the result
    out_img = np.dstack((binary_image, binary_image, binary_image))*255
    out_img = np.uint8(out_img)
    # Find the peak of the left and right halves of the histogram
    # These will be the starting point for the left and right lines
    midpoint = np.int(histogram.shape[0]/2)
    leftx_base = np.argmax(histogram[:midpoint])
    rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # Choose the number of sliding windows
    nwindows = 9
    # Set height of windows
    window_height = np.int(binary_image.shape[0]/nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_image.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated for each window
    leftx_current = leftx_base
    rightx_current = rightx_base
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 50
    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_image.shape[0] - (window+1)*window_height
        win_y_high = binary_image.shape[0] - window*window_height
        win_xleft_low = leftx_current - margin
        win_xleft_high = leftx_current + margin
        win_xright_low = rightx_current - margin
        win_xright_high = rightx_current + margin
        # Draw the windows on the visualization image
    #     print(win_xleft_low, win_y_low, win_xleft_high, win_y_high)
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),(win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),(win_xright_high,win_y_high),(0,255,0), 2) 
        # Identify the nonzero pixels in x and y within the window
        good_left_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xleft_low) & (nonzerox < win_xleft_high)).nonzero()[0]
        good_right_inds = ((nonzeroy >= win_y_low) & (nonzeroy < win_y_high) & (nonzerox >= win_xright_low) & (nonzerox < win_xright_high)).nonzero()[0]
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        # If you found > minpix pixels, recenter next window on their mean position
        if len(good_left_inds) > minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds) > minpix:        
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))

    # Concatenate the arrays of indices
    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)

    # Extract left and right line pixel positions
    left_line.allx = nonzerox[left_lane_inds]
    left_line.ally = nonzeroy[left_lane_inds] 
    right_line.allx = nonzerox[right_lane_inds]
    right_line.ally = nonzeroy[right_lane_inds] 

    # Fit a second order polynomial to each
    # and calculate radius of curvature
    y_eval = np.max(ploty)
    if len(left_line.allx) == 0:
        left_line.detected = False
    else:
        left_line.current_fit = np.polyfit(left_line.ally, left_line.allx, 2)
        left_line.detected = True
        left_fit_cr = np.polyfit(left_line.ally*ym_per_pix, left_line.allx*xm_per_pix, 2)
        left_line.radius_of_curvature = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    if len(right_line.allx) == 0:
        right_line.detected = False
    else:
        right_line.current_fit = np.polyfit(right_line.ally, right_line.allx, 2)
        right_line.detected = True
        right_fit_cr = np.polyfit(right_line.ally*ym_per_pix, right_line.allx*xm_per_pix, 2)
        right_line.radius_of_curvature = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    return left_line, right_line

def track_lanes(binary_image):
    global left_line
    global right_line
    global ym_per_pix
    global xm_per_pix
    ploty = np.linspace(0, binary_image.shape[0]-1, binary_image.shape[0] )
    nonzero = binary_image.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    margin = 100
    left_lane_inds = ((nonzerox > (left_line.current_fit[0]*(nonzeroy**2) + left_line.current_fit[1]*nonzeroy + left_line.current_fit[2] - margin)) & (nonzerox < (left_line.current_fit[0]*(nonzeroy**2) + left_line.current_fit[1]*nonzeroy + left_line.current_fit[2] + margin))) 
    right_lane_inds = ((nonzerox > (right_line.current_fit[0]*(nonzeroy**2) + right_line.current_fit[1]*nonzeroy + right_line.current_fit[2] - margin)) & (nonzerox < (right_line.current_fit[0]*(nonzeroy**2) + right_line.current_fit[1]*nonzeroy + right_line.current_fit[2] + margin)))  

    # Again, extract left and right line pixel positions
    left_line.allx = nonzerox[left_lane_inds]
    left_line.ally = nonzeroy[left_lane_inds] 
    right_line.allx = nonzerox[right_lane_inds]
    right_line.ally = nonzeroy[right_lane_inds]
    # Fit a second order polynomial to each
    y_eval = np.max(ploty)
    if len(left_line.allx) == 0:
        left_line.detected = False
    else:
        left_line.current_fit = np.polyfit(left_line.ally, left_line.allx, 2)
        left_line.detected = True
        left_fit_cr = np.polyfit(left_line.ally*ym_per_pix, left_line.allx*xm_per_pix, 2)
        left_line.radius_of_curvature = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    if len(right_line.allx) == 0:
        right_line.detected = False
    else:
        right_line.current_fit = np.polyfit(right_line.ally, right_line.allx, 2)
        right_line.detected = True
        right_fit_cr = np.polyfit(right_line.ally*ym_per_pix, right_line.allx*xm_per_pix, 2)
        right_line.radius_of_curvature = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    return left_line, right_line
    
```

```python
def draw_debug(warped_image):
    global left_line
    global right_line
    # Generate x and y values for plotting
    out_img = np.copy(warped_image)
    ploty = np.linspace(0, warped_image.shape[0]-1, warped_image.shape[0] )
    left_fitx = left_line.current_fit[0]*ploty**2 + left_line.current_fit[1]*ploty + left_line.current_fit[2]
    right_fitx = right_line.current_fit[0]*ploty**2 + right_line.current_fit[1]*ploty + right_line.current_fit[2]
    out_img[left_line.ally, left_line.allx] = [255, 0, 0]
    out_img[right_line.ally, right_line.allx] = [0, 0, 255]
    
    plt.plot(left_fitx, ploty, color='yellow')
    plt.plot(right_fitx, ploty, color='yellow')
    plt.xlim(0, 1280)
    plt.ylim(720, 0)

    return out_img

```

```python
ym_per_pix = 30/720 # meters per pixel in y dimension
xm_per_pix = 3.7/700 # meters per pixel in x dimension

image = mpimg.imread("left.jpg")
undistorted = cv2.undistort(image, mtx, dist, None, mtx)
binary_image = preprocess_img(undistorted)
binary_warped = warp(binary_image, M)
left_line = Line()
right_line = Line()
left_line, right_line = detect_lanes(binary_warped)

warped = warp(undistorted, M)
output = draw_debug(warped)
plt.imshow(output)
plt.show()
# plt.savefig('output_images/lane_marks.jpg')
```

# Lane offset

```python
def calculate_lane_offset(image):
    global left_line
    global right_line

    y = image.shape[0]-1
    left_line_position = left_line.current_fit[0]*y**2 + left_line.current_fit[1]*y + left_line.current_fit[2]
    right_line_position = right_line.current_fit[0]*y**2 + right_line.current_fit[1]*y + right_line.current_fit[2]
    lane_width_pix = right_line_position - left_line_position
    pix_to_x = 3.7 / lane_width_pix
    offset = image.shape[1] / 2 - (right_line_position - left_line_position) / 2
    return offset * pix_to_x
```

# Project back to camera image

```python
def prepare_warped_lane(image, left_line, right_line):
    ploty = np.linspace(0, image.shape[0]-1, image.shape[0] )
    left_fitx = left_line.current_fit[0]*ploty**2 + left_line.current_fit[1]*ploty + left_line.current_fit[2]
    right_fitx = right_line.current_fit[0]*ploty**2 + right_line.current_fit[1]*ploty + right_line.current_fit[2]

    warp_zero = np.zeros_like(image).astype(np.uint8)
    color_warp = warp_zero
    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0, 255, 0))
    return color_warp
    
def combine_images(base_img, overlay_img):
    result = cv2.addWeighted(base_img, 1, overlay_img, 0.3, 0)
    return result

```

```python
# continue from example at 'finding lane marks'


pnt1 = [30, 720]
pnt2 = [530, 465]
pnt3 = [750, 465]
pnt4 = [1250, 720]
points = [pnt1, pnt2, pnt3, pnt4]
src = np.array(points, np.float32)
dest = np.array([[0, 720], [0, 0], [1280, 0], [1280, 720]], np.float32)
M = cv2.getPerspectiveTransform(src, dest)
Minv = cv2.getPerspectiveTransform(dest, src)

lane_img = prepare_warped_lane(undistorted, left_line, right_line)
lane_unwarped = unwarp(lane_img, Minv)
result = combine_images(undistorted, lane_unwarped)
plt.imshow(result)
plt.show()
# plt.savefig('output_images/project_detected_lane.jpg')
```

```python

bad_index = 0
left_line = Line()
right_line = Line()
ym_per_pix = 30/720 # meters per pixel in y dimension
xm_per_pix = 3.7/700 # meters per pixel in x dimension

def lines_detected_correctly(num_lines = 720, debug = False):
    global left_line
    global right_line
    if left_line.detected == False | right_line.detected == False:
        print("left_line ({}) or right_line ({}) are False".format(left_line.detected, right_line.detected))
        return False
    # check distances are about the same at top, middle, bottom of image
#     ploty = np.array([0, image.shape[0] / 2, image.shape[0]])
    ploty = np.linspace(0, num_lines-1, 10)
    left_fitx = left_line.current_fit[0]*ploty**2 + left_line.current_fit[1]*ploty + left_line.current_fit[2]
    right_fitx = right_line.current_fit[0]*ploty**2 + right_line.current_fit[1]*ploty + right_line.current_fit[2]
    diff = right_fitx - left_fitx
    max = np.max(diff)
    min = np.min(diff)    
    mean = np.mean(diff)
    if debug == True:
        print("left_fitx: {}".format(left_fitx))
        print("right_fitx: {}".format(right_fitx))
        print("diff: {}".format(diff))
        print("max: {}, min: {} mean: {}".format(max, min, mean))
    if max - min > 510:
        print("diff is too large: {}".format(max-min))
        return False
    elif int(mean) not in range(600, 1050):
        print("mean in not in valid range {}".format(mean))
        return False
    else:
        return True

def process_image(image, debug=False):
    global left_line
    global right_line
    global bad_index
    undistorted = cv2.undistort(image, mtx, dist, None, mtx)
    binary_image = preprocess_img(undistorted)
    binary_warped = warp(binary_image, M)
    if lines_detected_correctly(debug) == True:
        left_line, right_line = track_lanes(binary_warped)
    if lines_detected_correctly(debug) == False:
        print("failed to track lane, do full detection")
        left_line, right_line = detect_lanes(binary_warped)

    if debug == True:
        print("orig image")
        plt.imshow(image)
        plt.show()
        print("binary_warped")
        plt.imshow(binary_warped)
        plt.show()
        warped = warp(undistorted, M)
        debug_output = draw_debug(warped)
        print("debug")
        plt.imshow(debug_output)
        plt.show()

    if lines_detected_correctly(debug) == False:
        bad_index += 1
        bad_imagename = "bad_image{}.jpg".format(bad_index)
        print("found bad image {}".format(bad_imagename))
        scipy.misc.imsave(bad_imagename, image)
        return undistorted
    
    lane_img = prepare_warped_lane(undistorted, left_line, right_line)
    lane_unwarped = unwarp(lane_img, Minv)
    result = combine_images(undistorted, lane_unwarped)
    if result is None:
        print("failed to produce image")
        return undistorted
    else:
        font = cv2.FONT_HERSHEY_SIMPLEX
        radius = (left_line.radius_of_curvature + right_line.radius_of_curvature) / 2
        cv2.putText(result, "Radius: {0:.2f}m".format(radius) ,(50, 50), font, 1, (0,0,0),2,cv2.LINE_AA)
        lane_offset = calculate_lane_offset(result)
        cv2.putText(result, "Lane offset: {0:.2f}m".format(lane_offset) ,(50, 85), font, 1, (0,0,0),2,cv2.LINE_AA)
        return result


video_filename_prefix = "project_video"
# video_filename_prefix = "challenge_video"
clip1 = VideoFileClip("{}.mp4".format(video_filename_prefix))
white_clip = clip1.fl_image(process_image) #NOTE: this function expects color images!!
%time white_clip.write_videofile("{}_output.mp4".format(video_filename_prefix), audio=False)
```
